```python
import pandas as pd
import pingouin as pg
import seaborn as sns
import matplotlib.pyplot as plt

# Set visualization style for academic plots
sns.set_theme(style="whitegrid", palette="muted")
```

```python
# Load the cleaned behavioral and survey datasets
# Update these paths if your files are in a different folder (e.g., 'data/processed/...')
behavioral_df = pd.read_csv('behavioral_events_master.csv')
survey_df = pd.read_csv('HCI-Jazz_Cleaned_Responses.csv')

# Merge the two datasets
# Behavioral uses 'Participant' and Survey uses 'participant_id'
merged_df = pd.merge(
    behavioral_df, 
    survey_df, 
    left_on='Participant', 
    right_on='participant_id', 
    how='inner'
)

# Verify the merge was successful
print(f"Merged Dataset Shape: {merged_df.shape}")
print("Sample of merged groups and conditions:")
print(merged_df[['Participant', 'group', 'Condition', 'alarm_file', 'reaction_time_ms']].head())
```

```python
# Group the data to get the mean reaction time per participant per condition
# This yields one 'Consonant' average and one 'Dissonant' average per user
agg_df = merged_df.groupby(['Participant', 'group', 'Condition'])['reaction_time_ms'].mean().reset_index()

# Display the descriptive statistics (Means and Standard Deviations)
desc_stats = agg_df.groupby(['group', 'Condition'])['reaction_time_ms'].agg(['mean', 'std', 'count']).round(2)
print("--- Descriptive Statistics ---")
print(desc_stats)
```

```python
plt.figure(figsize=(8, 6))

# Create an interaction plot
sns.pointplot(
    data=agg_df, 
    x='Condition', 
    y='reaction_time_ms', 
    hue='group',
    markers=["o", "s"],
    capsize=.1,
    err_kws={'linewidth': 1.5},
    dodge=True
)

plt.title('Interaction Effect: Alarm Condition vs. Musical Expertise', fontsize=14, pad=15)
plt.ylabel('Average Reaction Time (ms)', fontsize=12)
plt.xlabel('Harmonic Condition', fontsize=12)
plt.legend(title='Expertise Group')
plt.tight_layout()

# Save the plot for your thesis
plt.savefig('interaction_plot.png', dpi=300)
plt.show()
```
